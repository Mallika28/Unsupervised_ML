```python
data=df.iloc[:,[1,2,3,4,5]].values
data
```

```python
data = df.iloc[:, [0, 1, 2, 3,4]].values

from sklearn.cluster import KMeans
mallika = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(data)
    mallika.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), mallika)
plt.title('The elbow method',fontsize=20)
plt.xlabel('Number of clusters')
plt.ylabel('mallika') # Within cluster sum of squares
plt.show()
```

```python

kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(data)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(data[y_kmeans == 0, 0], data[y_kmeans == 0, 1], s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(data[y_kmeans == 1, 0], data[y_kmeans == 1, 1], s = 100, c = 'skyblue', label = 'Iris-setosa')
plt.scatter(data[y_kmeans == 2, 0], data[y_kmeans == 2, 1], s = 100, c = 'orange', label = 'Iris-virginica');
plt.grid()
```

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'green', label = 'Centroids')
plt.legend();
plt.grid()
```
